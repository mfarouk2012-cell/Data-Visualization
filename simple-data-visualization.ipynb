# **Simple Data Visualization**

# **Required libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statistics as st
```

# **USER CUSTOMIZATION SECTION**

```python
x= categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5', 'Category 6']
```

```python
y= values_simple = [25, 40, 35, 50, 18, 12]
```

**For Pie**

```python
explode = [0, 0, 0, 0.1, 0, 0]
```

```python
chart_title = 'Your Chart Title Here'
x_axis_label = 'Your X-Axis Label Here'
y_axis_label = 'Your Y-Axis Label Here'
```

chart title color:

```python
color1= color="#DF204A"
```

x_axis_label color:

```python
color2= color="#6320DF"
```

y_axis_label color:

```python
color3= color="#0BBD73"
```

```python
colors = [
    "#558EEB", # Blue
    "#B8DA21", # Orange
    '#55A868', # Green
    "#DF3035", # Red
    '#8172B2', # Purple
     
]
```

# **Calculations:**

number of items

```python
items_no = len(y)
```

max value

```python
max_value= max(y)
max_index = y.index(max_value) 
max_category = x[max_index]
```

total

```python
total_sum = sum(y)
```

min value

```python
min_value= min(y)
min_index = y.index(min_value) 
min_category = x[min_index]
```

Mean

```python
mean_value = np.mean(y)
```

Median

```python
median_value = np.median(y)
```

Mode

```python
mode_value = st.mode(y)
```

# **Printing statistical calculations**


```python
def statistical_calculations():
 print ( "Number of elements = " + str(items_no)) 
 print("Total_sum = ", (total_sum))
 print("Max_ value = ", (max_value))
 print("Category for Max_ value = ", (max_category))

 print("Min_ value = ", (min_value))
 print("Category for Min_ value = ", (min_category))

 print("The mean = ", (mean_value)) 
 print("The median = ", (median_value))   
 print("The mode = ", (mode_value))   
```

# **(1) Line Chart (Line Plot)** 

```python

x= categories
y= values_simple
plt.title(chart_title, fontweight='bold', fontsize=14, pad=20, color= color1)
plt.xlabel(x_axis_label, color= color2)
plt.ylabel(y_axis_label, color= color3)
plt.grid(True, linestyle='--', alpha=0.6)

plt.plot(x, y, color= "#1E1149")
plt.show()

# printing statistical calculations
statistical_calculations()
```

# **(2) Scatter plot**

```python
x = np.array(categories)
y = np.array(values_simple)
x_numeric = np.arange(len(categories))

plt.scatter(x_numeric, y, marker='.', s=150, color="#E0296F")
m, c = np.polyfit(x_numeric, y, 1)

plt.plot(x_numeric, m * x_numeric + c, color='#8172B2')
plt.xticks(x_numeric, x)

plt.title(chart_title, fontweight='bold', fontsize=14, pad=20, color= color1)
plt.xlabel(x_axis_label, color= color2)
plt.ylabel(y_axis_label, color= color3)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()

# printing statistical calculations
statistical_calculations()

```

# **(3) Bar chart**

```python
x = categories
y = values_simple


#plt.figure(figsize=(15, 6))

plt.bar(
    x,                      
    y,             
    align='center',
    alpha=0.9,
    color=colors

)

plt.title(chart_title, fontweight='bold', fontsize=14, pad=20, color= color1)
plt.xlabel(x_axis_label, color= color2)
plt.ylabel(y_axis_label, color= color3)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()

# printing statistical calculations
statistical_calculations()

```

# **(4) Pie chart**

```python
plt.pie(y, labels=x, explode=explode)

chart_title = 'Your Chart Title Here'

plt.title(chart_title, fontweight='bold', fontsize=14, pad=20, color= color1)

plt.show()

# printing statistical calculations
statistical_calculations()
```

# **All charts together**

```python
# Create a single figure with 4 subplots
plt.figure(figsize=(20, 20))  # Wider figure to accommodate 4 plots

# Subplot 1: Line plot
plt.subplot(2, 2, 1)
plt.plot(x, y, color="#1E1149")
plt.title(chart_title, fontweight='bold', fontsize=14, pad=20, color=color1)
plt.xlabel(x_axis_label, color=color2)
plt.ylabel(y_axis_label, color=color3)
plt.grid(True, linestyle='--', alpha=0.6)

# Subplot 2: Scatter plot with regression line
plt.subplot(2, 2, 2)
plt.scatter(x_numeric, y, marker='.', s=150, color="#E0296F")
m, c = np.polyfit(x_numeric, y, 1)
plt.plot(x_numeric, m * x_numeric + c, color='#8172B2')
plt.xticks(x_numeric, x)
plt.title(chart_title, fontweight='bold', fontsize=14, pad=20, color=color1)
plt.xlabel(x_axis_label, color=color2)
plt.ylabel(y_axis_label, color=color3)
plt.grid(True, linestyle='--', alpha=0.6)

# Subplot 3: Bar chart
plt.subplot(2, 2, 3)
plt.bar(x, y, align='center', alpha=0.9, color=colors)
plt.title(chart_title, fontweight='bold', fontsize=14, pad=20, color=color1)
plt.xlabel(x_axis_label, color=color2)
plt.ylabel(y_axis_label, color=color3)
plt.grid(True, linestyle='--', alpha=0.6)

# Subplot 4: Pie chart
plt.subplot(2, 2, 4)
plt.pie(y, labels=x, explode=explode)
plt.title(chart_title, fontweight='bold', fontsize=14, pad=20, color=color1)

# Adjust layout and show ALL plots together
plt.tight_layout()
plt.show()

# Printing statistical calculations
statistical_calculations()
```
